```python
import os
import numpy as np
import pandas as pd
from biom import Table, load_table
from biom.util import biom_open
```

```python
results_dir = '../../data/simulations/soil_benchmarking/'
input_dir = '../../data/case-studies/biocrust/'
microbes = load_table(os.path.join(input_dir, 'microbes.biom'))
metabolites = load_table(os.path.join(input_dir, 'metabolites.biom'))
shared_ = set(microbes.ids()) & set(metabolites.ids())
microbes = microbes.filter(shared_)
metabolites = metabolites.filter(shared_)
mf = pd.DataFrame([x.split('_')[0].replace('p5','') 
                   for x in metabolites.ids()], metabolites.ids(), ['time'])
mf['train_test'] = 'train'
mf.loc[['mid' in x for x in mf.index], 'train_test'] = 'test'
mf.index.name = 'sample_name'
#mf.to_csv(os.path.join(results_dir, 'data/metadata.tsv'), sep='\t')
(microbes, metabolites)
```

```python
"""
subsample the tables
"""

density_= 0
coeff_microbe = 270
coeff_metabolite = 1 * 1e5

# subsample microbes
rel_microbe_subsampled = pd.DataFrame(microbes.matrix_data.toarray(),
                                      microbes.ids('observation'),
                                      microbes.ids())
for s in rel_microbe_subsampled.columns:
    probs = (rel_microbe_subsampled[s] / sum(rel_microbe_subsampled[s])).values
    depth = sum(rel_microbe_subsampled[s]) // coeff_microbe
    dist_ = np.random.multinomial(depth, probs)
    rel_microbe_subsampled[s] =dist_
rel_microbe_subsampled.index.name ='OTU_ID'
#rel_microbe_subsampled.to_csv(os.path.join(results_dir, 'data/rel.microbes.%i.tsv' % (density_)), sep='\t')
rel_microbe_subsampled = Table(rel_microbe_subsampled.values,
                               rel_microbe_subsampled.index,
                               rel_microbe_subsampled.columns)   

# subsample metabolites
rel_metabolites_subsampled = pd.DataFrame(metabolites.matrix_data.toarray(),
                                              metabolites.ids('observation'),
                                              metabolites.ids())
for s in rel_metabolites_subsampled.columns:
    # calculate probs
    probs = (rel_metabolites_subsampled[s] / sum(rel_metabolites_subsampled[s])).values
    depth = sum(rel_metabolites_subsampled[s]) // coeff_metabolite
    dist_ = np.random.multinomial(depth, probs)
    rel_metabolites_subsampled[s] = dist_

rel_metabolites_subsampled.index.name ='OTU_ID'
#rel_metabolites_subsampled.to_csv(os.path.join(results_dir, 'data/rel.metabolites.%i.tsv' % (density_)), sep='\t')
rel_metabolites_subsampled = Table(rel_metabolites_subsampled.values, 
                                      rel_metabolites_subsampled.index,
                                      rel_metabolites_subsampled.columns)

print(np.mean(rel_microbe_subsampled.sum('sample')))
(rel_microbe_subsampled, rel_metabolites_subsampled)
```
